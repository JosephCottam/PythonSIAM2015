## [bokeh.charts](http://bokeh.pydata.org/docs/user_guide/charts.html)
Common Schemas for common tasks (and parameters).

Expects data to be formatted as either an OrderedDict or a pandas dataframe.

Supported Schemas: Bar, Boxplot, Categorical Heatmap, Histogram, Scatter, Timeseries

```python
from collections import OrderedDict
from bokeh.charts import Histogram

mu, sigma = 0, 0.5
normal_dist = OrderedDict(normal=np.random.normal(mu, sigma, size=1000))
hist = Histogram(normal_dist, bins=50,
                 title="kwargs, dict_input", 
                 ylabel="frequency", 
                 legend="top_left",
                 width=400, 
                 height=350, 
                 notebook=True)
hist.show()
```

```python
## Chained
# setting mu and sigma give the pdf and cdf of the idealized normal distribution with those parameters
hist = Histogram(normal_dist, bins=50, mu=mu, sigma=sigma).\
            title("kwargs, dict_input").\
            ylabel("frequency").\
            legend("top_left").\
            width(600).\
            height(300).\
            notebook(True)
hist.show()
```

```python
from bokeh.sampledata.olympics2014 import data
from bokeh.charts import Bar

data = {d['abbr']: d['medals'] for d in data['data'] if d['medals']['total'] > 0}

countries = sorted(data.keys(), key=lambda x: data[x]['total'], reverse=True)

gold = np.array([data[abbr]['gold'] for abbr in countries], dtype=np.float)
silver = np.array([data[abbr]['silver'] for abbr in countries], dtype=np.float)
bronze = np.array([data[abbr]['bronze'] for abbr in countries], dtype=np.float)

medals = OrderedDict(bronze=bronze, silver=silver, gold=gold)
bar = Bar(medals, countries, title="grouped, dict_input", xlabel="countries", ylabel="medals", 
          legend=True, width=800, height=300, notebook=True)
bar.show()
```

```python
from bokeh.sampledata.iris import flowers
from bokeh.charts import Scatter

setosa = flowers[(flowers.species == "setosa")][["petal_length", "petal_width"]]
versicolor = flowers[(flowers.species == "versicolor")][["petal_length", "petal_width"]]
virginica = flowers[(flowers.species == "virginica")][["petal_length", "petal_width"]]

split_flowers = OrderedDict([("setosa", setosa.values), ("versicolor", versicolor.values), ("virginica", virginica.values)])

scatter = Scatter(split_flowers)
scatter.title("iris dataset, dict_input").xlabel("petal_length").ylabel("petal_width").legend("top_left").width(600).height(400).notebook()

scatter.show()
```

```python
from bokeh.sampledata import us_counties, unemployment
from bokeh.plotting import *

county_xs=[
    us_counties.data[code]['lons'] for code in us_counties.data
    if us_counties.data[code]['state'] == 'tx'
]
county_ys=[
    us_counties.data[code]['lats'] for code in us_counties.data
    if us_counties.data[code]['state'] == 'tx'
]

colors = ["#F1EEF6", "#D4B9DA", "#C994C7", "#DF65B0", "#DD1C77", "#980043"]
```

```python
county_colors = []
for county_id in us_counties.data:
    if us_counties.data[county_id]['state'] != 'tx':
        continue
    try:
        rate = unemployment.data[county_id]
        idx = min(int(rate/2), 5)
        county_colors.append(colors[idx])
    except KeyError:
        county_colors.append("black")
```

```python
figure()
patches(county_xs, county_ys, fill_color=county_colors, fill_alpha=0.7,
        line_color="white", line_width=0.5, title="Texas Unemployment 2009")
show()
```

## Bokeh Server
Dynamic data & big data

Start the server
* Start a command line

* execute `bokeh-server`

[Official model description](http://bokeh.pydata.org/docs/user_guide/server.html)

**Note:** Server support for ipython notebook is still in development.  Some unusual effects occur when mixing server and non-server plots in one notebook.  Examples can also be run from the command line in bokeh/examples/plotting/server.

### Serving plots

```python
import time
from numpy import pi, cos, sin, linspace, roll, zeros_like
from bokeh.plotting import *
from bokeh.objects import Glyph

N = 50 + 1
r_base = 8
theta = linspace(0, 2*pi, N)
r_x = linspace(0, 6*pi, N-1)
rmin = r_base - cos(r_x) - 1
rmax = r_base + sin(r_x) + 1
colors = ["FFFFCC", "#C7E9B4", "#7FCDBB", "#41B6C4", "#2C7FB8", "#253494", "#2C7FB8", "#41B6C4", "#7FCDBB", "#C7E9B4"] * 5
cx = cy = zeros_like(rmin)

output_notebook(url="default")
figure(x_range=[-11, 11], y_range=[-11, 11], tools="pan,wheel_zoom,box_zoom,reset,previewsave")
annular_wedge(
    cx, cy, rmin, rmax, theta[:-1], theta[1:],
    inner_radius_units="data",
    outer_radius_units="data",
    fill_color = colors,
    line_color="black",
)
```

### Server-side downsample

Sometimes the network pipe is too small.  Bokeh server can perform visualization-aware server-side data manipulation to just send the required data.  Example files are in bokeh/examples/plotting/server.  Relevant files are remote_image.py, serversource.py, abstractrender.py and census.py.

Additional software must be installed:
* abstract_rendering -- Enables server-side rendering and data binning
* array_management -- Data access library (will be replaced by Blaze soon)

Start the server with additional flags:  bokeh-server -D <data directory>
